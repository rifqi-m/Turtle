```python
#Linear Regression with substitution
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

# data creation
np.random.seed(1)    
x= np.random.rand(100,1)*10
y= 2*x + 5 + np.random.randn(100,1)    # related to .coef_, .intercept_
                                       # 2*x with .coef_
                                       # 5 with .intercept_

# model creation, training 
model = linear_model.LinearRegression()
model.fit(x,y)

print(model.coef_)    # weights ,resulting 1.96 near 2, that means resulting good value 
print(model.intercept_)    # 2 number behind noice (bias) ,resulting 5.23 near 5, that means resulting good value 

# line creation
x_plot=np.arange(0,11)
x_plot=x_plot.reshape(-1,1)
y_predicted=model.predict(x_plot)

plt.figure()
plt.scatter(x,y)
plt.plot(x_plot,y_predicted,color='red') # line
plt.show()
```

```python
# Linear Regression with substitution
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
import pandas as pd

# data creation
np.random.seed(1)
x= np.random.randn(100,1)*3
z= 40 + x + 2*(x**2) +   np.random.randn(100,1)*5    # related to .coef_, .intercept_
                                                    # 2*x, x with .coef_
                                                    # 5 with .intercept_

# model creation, training
model = linear_model.LinearRegression()
x2=(x**2)

data=pd.DataFrame(x,columns=['x'])
data['x2']=x2
model.fit(data,z)

print(model.coef_)    # weights ,resulting 1.1 near 1 and 2.008 near 2, that means resulting good value
print(model.intercept_)    # 2 number behind noice (bias) ,resulting 40.6 near 40, that means resulting good value 


# line creation
x_plot=np.linspace(-8,8,100)
plot_data=pd.DataFrame(x_plot,columns=['x'])    #
plot_data['x2']=x_plot**2    #
z_predicted=model.predict(plot_data)    #

plt.figure()
plt.scatter(x,z)
plt.plot(x_plot,z_predicted,color='red')    # line
plt.show()
```

```python
#Higher degree polynomials with numpy
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

# data creation 
np.random.seed(1)
x= np.random.rand(20)*10
y= x + 5 + np.random.randn(20)

p5 = np.poly1d(np.polyfit(x, y, 5))    # coefficient
print(p5)

# 
plt.plot([i for i in range(10)],  [p5(i) for i in range(10)], c = 'green')
plt.scatter(x,y)
plt.show()

#
plt.scatter(x,y)
plt.plot(np.linspace(0, 9, 100),  p5(np.linspace(0, 9, 100)), c="red")
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

```python
#Overfitting
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

np.random.seed(1)
x=np.random.rand(20)*20
y=x**2 + 2*x + np.random.normal(loc=0.0, scale=100.0, size=20)
#weird size tuple is to ensure that addition with (20,1) matrix x works

plt.scatter(x,y,c = 'black')
plt.ylim(-100,800)
start=min(x)
end=max(x)

# polynomial
p = np.poly1d(np.polyfit(x, y, 3))    # 3 coefficient
plt.plot(np.linspace(start, end, 100),p(np.linspace(start, end, 100)),c="red")

p = np.poly1d(np.polyfit(x, y, 5))     # 5 coefficient
plt.plot(np.linspace(start, end, 100),p(np.linspace(start, end, 100)),c="blue")

p = np.poly1d(np.polyfit(x, y, 10))     # 10 coefficient
plt.plot(np.linspace(start, end, 100),p(np.linspace(start, end, 100)),c="green")

p = np.poly1d(np.polyfit(x, y, 20))     # 20 coefficient 
plt.plot(np.linspace(start, end, 100),p(np.linspace(start, end, 100)),c="cyan")
plt.show()
```

```python
#sample Solution (only with pandas)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

# data input
data = pd.read_csv("houses.csv")
data.columns = ['living_space','size_of_property','price']    # create column names

#corr = data.corr()

# numpy method
degree = 5

# create model training
model = np.poly1d(np.polyfit(data.living_space, data.price, degree))
predicted_prices = model(data.living_space)    # training model


print_data = np.linspace(0, np.max(data.living_space),100)

plt.figure()
plt.scatter(data.living_space, data.price)    # real data
plt.plot(print_data, model(print_data),c="black")    # line, model data
plt.ylim(min(data.price)*0.8, max(data.price)*1.2)    # y axes
plt.show()

#MAE
err = np.mean(np.abs(predicted_prices-data.price))
print(f'The MAE is {err}')


house280 = model(280)
print(f'The prediction for a house of size 280 is {house280}')
```

```python
# Differently Complex Polynoms and Errors   (on Test Data)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

data = pd.read_csv("houses.csv",header=None)
data.columns = ['living_space','size_of_property','price']

test_data = pd.read_csv("houses_test.csv",header=None)
test_data.columns = ['living_space','size_of_property','price']

# plot creation in each polynomyal/ degree
for degree in range(1,11):
    model = np.poly1d(np.polyfit(data.living_space, data.price, degree))
    predicted_prices = model(test_data.living_space)

    print_data = np.linspace(0, np.max(test_data.living_space),100)
    plt.figure()
    plt.scatter(data.living_space, data.price,c="green")
    plt.scatter(test_data.living_space, test_data.price,c="red")
    plt.plot(print_data, model(print_data), c="black")
    plt.ylim(min(data.price)*0.8,max(data.price)*1.2)
    plt.show()
    
    err = np.mean(np.abs(predicted_prices-test_data.price))
    print(f'The MAE for degree {degree:2} is {err:11.3f}')
```

```python
# solution   (exercise)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

data = pd.read_csv("houses.csv",header=None)
data.columns = ['living_space','size_of_property','price']
test_data = pd.read_csv("houses_test.csv",header=None)
test_data.columns = ['living_space','size_of_property','price']

model = linear_model.LinearRegression()
model.fit(data[['living_space','size_of_property']],data.price)
predicted_prices = model.predict(test_data[['living_space','size_of_property']])

err = np.mean(np.abs(predicted_prices-test_data.price))
print(f'The MAE is {err:11.3f}')
print(f'coefficients: {model.coef_}')
print(f'bias: {model.intercept_}')
```
